Line Chart: Trend of Bird Strikes Over Time


```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Select features and target
features = df1[['Altitude bin', 'Aircraft: Type', 'Conditions: Precipitation', 'Conditions: Sky', 'When: Phase of flight']]
target = df1['Wildlife: Number Struck Actual']

# One-hot encode categorical features
features_encoded = pd.get_dummies(features, drop_first=True)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(features_encoded, target, test_size=0.2, random_state=42)

# Random Forest Regressor
rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, y_train)
rf_predictions = rf.predict(X_test)

# Evaluate the model
mae = mean_absolute_error(y_test, rf_predictions)
rmse = mean_squared_error(y_test, rf_predictions, squared=False)

print(f'Random Forest MAE: {mae}')
print(f'Random Forest RMSE: {rmse}')
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(y_test, rf_predictions, alpha=0.5, label='Predicted')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', label='Actual')
plt.xlabel('Actual Number of Strikes')
plt.ylabel('Predicted Number of Strikes')
plt.title('Actual vs Predicted Number of Bird Strikes')
plt.legend()
plt.show()
```

Does the phase of flight affect the likelihood of wildlife strikes?

```python
# Calculate total strikes per phase of flight
phase_strikes_summary = df1.groupby('When: Phase of flight').size().reset_index(name='Total Strikes')

# Calculate the proportion of strikes for each phase
phase_strikes_summary['Proportion'] = phase_strikes_summary['Total Strikes'] / phase_strikes_summary['Total Strikes'].sum()

print(phase_strikes_summary)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot total strikes per phase of flight
plt.figure(figsize=(12, 6))
sns.barplot(
    x='When: Phase of flight',
    y='Total Strikes',
    data=phase_strikes_summary,
    palette='Set2'
)
plt.xlabel('Phase of Flight')
plt.ylabel('Total Number of Strikes')
plt.title('Total Number of Wildlife Strikes per Phase of Flight')
plt.xticks(rotation=45)
plt.show()

# Plot proportions
plt.figure(figsize=(12, 6))
sns.barplot(
    x='When: Phase of flight',
    y='Proportion',
    data=phase_strikes_summary,
    palette='Set2'
)
plt.xlabel('Phase of Flight')
plt.ylabel('Proportion of Total Strikes')
plt.title('Proportion of Wildlife Strikes per Phase of Flight')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot pie chart of strikes by phase of flight
plt.figure(figsize=(8, 8))
plt.pie(
    phase_strikes_summary['Total Strikes'],
    labels=phase_strikes_summary['When: Phase of flight'],
    autopct='%1.1f%%',
    colors=sns.color_palette('Set2', len(phase_strikes_summary)),
    startangle=140
)
plt.title('Distribution of Wildlife Strikes by Phase of Flight')
plt.show()
```

```python
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df1['When: Phase of flight'], 'Count')

# Perform Chi-Square test
chi2, p, _, _ = chi2_contingency(contingency_table)

print(f'Chi-Square Statistic: {chi2}')
print(f'p-value: {p}')

# Interpretation of p-value
if p < 0.05:
    print("There is a significant association between phase of flight and the number of wildlife strikes.")
else:
    print("There is no significant association between phase of flight and the number of wildlife strikes.")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Select features and target
features = df1[['Altitude bin', 'Aircraft: Type', 'Conditions: Precipitation', 'Conditions: Sky', 'When: Phase of flight']]
target = df1['Wildlife: Number Struck Actual']

# One-hot encode categorical features
features_encoded = pd.get_dummies(features, drop_first=True)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(features_encoded, target, test_size=0.2, random_state=42)

# Linear Regression
lr = LinearRegression()
lr.fit(X_train, y_train)
lr_predictions = lr.predict(X_test)
print(f'Linear Regression MAE: {mean_absolute_error(y_test, lr_predictions)}')
print(f'Linear Regression RMSE: {mean_squared_error(y_test, lr_predictions, squared=False)}')

# Random Forest Regressor
rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, y_train)
rf_predictions = rf.predict(X_test)
print(f'Random Forest MAE: {mean_absolute_error(y_test, rf_predictions)}')
print(f'Random Forest RMSE: {mean_squared_error(y_test, rf_predictions, squared=False)}')
```
